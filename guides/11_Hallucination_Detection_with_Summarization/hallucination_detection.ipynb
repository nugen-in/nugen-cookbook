## **Hallucination Detection with Summarization**

This notebook demonstrates how to summarize domain-specific text and detect potential hallucinations in the generated summary using Nugen's API.

### **Nugen API Integration**

**Set up the Nugen API Key**

```python
import os
import requests
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("NUGEN_API_KEY")
if not API_KEY:
    raise EnvironmentError("NUGEN_API_KEY environment variable not set.")

BASE_URL = "https://api.nugen.in/api/v3/inference/completions"
MODEL = "nugen-flash-instruct"

headers = 
    "Authorization": f"Bearer {API_KEY}",
    "Content-Type": "application/json"
}

print("API configured successfully")
```

### **Load Source Document**

Sample healthcare text for summarization

```python
SOURCE_DOCUMENT = """
Diabetes mellitus is a chronic metabolic disorder characterized by elevated blood glucose levels 
resulting from defects in insulin secretion, insulin action, or both. Type 1 diabetes accounts 
for approximately 5-10% of all diabetes cases and is caused by autoimmune destruction of 
pancreatic beta cells. Type 2 diabetes, representing 90-95% of cases, is primarily associated 
with insulin resistance and relative insulin deficiency. Common symptoms include polyuria, 
polydipsia, unexplained weight loss, and fatigue. Long-term complications can affect the eyes 
(retinopathy), kidneys (nephropathy), nerves (neuropathy), and cardiovascular system.
"""

print("Source Document:")
print(SOURCE_DOCUMENT.strip())
```

### **Generate Summary**

Use Nugen API to create a concise summary

```python
summary_prompt = f"""Summarize the following medical text in 2-3 sentences. Focus on key facts.

Text: {SOURCE_DOCUMENT}

Summary:"""

payload = {
    "model": MODEL,
    "prompt": summary_prompt,
    "max_tokens": 150,
    "temperature": 0.3
}

response = requests.post(BASE_URL, json=payload, headers=headers, timeout=30)

if response.status_code == 200:
    summary = response.json()["choices"][0]["text"].strip()
    print("Generated Summary:")
    print(summary)
else:
    print(f"Error {response.status_code}: {response.text}")
    summary = None
```

### **Detect Hallucinations**

Verify if the summary contains facts not present in the source

```python
if summary:
    # Use a direct question format
    verify_prompt = f"""Original text: {SOURCE_DOCUMENT}

Summary: {summary}

Question: Does the summary accurately reflect the facts in the original text? Answer YES or NO and explain why."""

    payload = {
        "model": MODEL,
        "prompt": verify_prompt,
        "max_tokens": 150,
        "temperature": 0.3
    }

    response = requests.post(BASE_URL, json=payload, headers=headers, timeout=30)

    if response.status_code == 200:
        verification = response.json()["choices"][0]["text"].strip()
        print("Hallucination Check:")
        print(verification)
    else:
        print(f"Error {response.status_code}: {response.text}")
```

### **Display Results**

Summary and hallucination check side-by-side

```python
print("=" * 60)
print("SUMMARIZATION & HALLUCINATION REPORT")
print("=" * 60)
print("\n[SOURCE]")
print(SOURCE_DOCUMENT.strip()[:200] + "...")
print("\n[SUMMARY]")
print(summary if summary else "Failed to generate")
print("\n[VERIFICATION]")
print(verification if 'verification' in dir() else "Not available")
print("=" * 60)
```
