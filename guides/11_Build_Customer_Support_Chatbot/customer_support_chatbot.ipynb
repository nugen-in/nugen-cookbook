```python
import requests
import json
import time
```

## Setup

Get your API key from [platform.nugen.in](https://platform.nugen.in)

```python
API_KEY = "<your-api-key-here>"  # Replace with your key
BASE_URL = "https://api.nugen.in/agents"

headers = {
    "Authorization": f"Bearer {API_KEY}",
    "Content-Type": "application/json"
}
```

## Create the Support Agent

```python
agent_config = {
    "agent_name": "TechStore Support",
    "agent_description": "Customer support for TechStore Electronics",
    "model": "llama-v3p3-70b-instruct",
    "temperature": 0.3,
    "instructions": """You are a friendly support agent for TechStore Electronics.
Help customers with product questions, orders, returns, and basic troubleshooting.
Be concise and helpful. Our policies: 30-day returns for unopened items, 
14-day for opened items, free shipping over $50.""",
    "demonstrations": [
        {
            "user_input": "What's your return policy?",
            "model_output": "We offer 30-day returns for unopened items and 14-day for opened items. Just bring your receipt!"
        },
        {
            "user_input": "My laptop won't turn on",
            "model_output": "Try this: 1) Check it's plugged in, 2) Hold power button 15 seconds, 3) Remove battery if possible. Did any help?"
        }
    ]
}

response = requests.post(f"{BASE_URL}/create/", json=agent_config, headers=headers)
agent = response.json()
AGENT_ID = agent['id']
print(f"Agent created: {AGENT_ID}")
```

## Chat with the Bot

```python
def chat(message):
    response = requests.post(
        f"{BASE_URL}/{AGENT_ID}/run/",
        json={"message": message},
        headers=headers
    )
    return response.json().get("response", "Error")
```

```python
# Test it out
questions = [
    "Do you have iPhones in stock?",
    "What's your return policy?",
    "My Bluetooth headphones won't connect"
]

for q in questions:
    print(f"Customer: {q}")
    print(f"Bot: {chat(q)}\n")
    time.sleep(1)
```

## Try Your Own Questions

```python
print(chat("Can I get a price match?"))
```

```python
print(chat("Where is my order TS-12345?"))
```

## Usage Stats

```python
stats = requests.get(f"{BASE_URL}/usage/", headers=headers).json()
print(json.dumps(stats, indent=2))
```
