```python
import os
import json
import requests

from dotenv import load_dotenv
```

```python
load_dotenv()

NUGEN_API_KEY = os.getenv("NUGEN_API_KEY")

print("API Key Loaded Successfully")
```

```python
BASE_URL = "https://api.nugen.in"

CHAT_ENDPOINT = f"{BASE_URL}/api/v3/inference/completions"

HEADERS = {
    "Authorization": f"Bearer {NUGEN_API_KEY}",
    "Content-Type": "application/json"
}

PLANNER_MODEL = "Llama V3P2 3B Reasoning"
```

## Create Planning Prompt

The planning agent converts user goals into structured execution steps.

```python
planning_prompt = """
You are a planning agent.

Your task is to analyze the user query and break it down into a clear, logical, step-by-step plan.

Rules:
- Do NOT answer directly
- Return only a structured plan
- Use numbered steps
- Keep steps concise and actionable

User Request:
Build and launch an AI-powered creator assistant using Nugen APIs.
"""

payload = {
    "model": PLANNER_MODEL,
    "prompt": planning_prompt
}

response = requests.post(
    CHAT_ENDPOINT,
    headers=HEADERS,
    json=payload
)

planning_result = response.json()

print(json.dumps(planning_result, indent=2))

generated_plan = planning_result["choices"][0]["text"]

print("\n===== Generated Execution Plan =====\n")

print(generated_plan)
```
